````python
# For more information -> https://www.folkstalk.com/2022/09/sql-server-python-connection-with-code-examples.html
#create database in python with mysql

def runSQL_Query(sql_query, databaseCreated):
  # Database connection string  
  config = {
    'user': 'root',
    'password': 'Mixtli666',
    'host': '127.0.0.1',
    'raise_on_warnings': True,
    'database' : "employees"
  }
  try:    
    mydb = mysql.connector.connect(**config)
    mycursor = mydb.cursor()
    if(databaseCreated):
      mycursor.execute("Use employees;")  
    mycursor.execute(sql_query)
    mydb.commit()   

  except mysql.connector.Error as err:
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
      print("Something is wrong with your user name or password")
    elif err.errno == errorcode.ER_BAD_DB_ERROR:
      print("Database does not exist")
    else:
      print(err)
  finally:
    mycursor.close()
    mydb.close()

# SQL principal que crea la BD, tablas y 2 vistas
file_zip_db = os.path.join(folder_unziped, "employees.sql")
file_sql = open(file_zip_db, "r")
sql_query_emp = file_sql.read()

# Run query to create database
runSQL_Query(sql_query_emp, False)
````

## Creando vistas

### ¿Que son las vistas?
Son tablas creadas a partir de otra u otras tablas, estas no contien los datos, solo mandan a llamar a otra u otra tablas cuando son ejecutadas.

### ¿Como crear una vista de los empleados?
En este caso crearemos una vista con toda la información de los empleados, asi sera mas facil accesar a la información cuando la necesitemos.
Ejemplo:
```USE employees;
CREATE  OR REPLACE VIEW v_All_Employee_Data AS
SELECT e.emp_no, birth_date,first_name, last_name, e.gender, e.hire_date, d.dept_no, d.dept_name, de.from_date as dept_from_Date, de.to_date as dept_to_date, t.title, t.from_date, t.to_date
 FROM employees e
inner join dept_emp de on de.emp_no = e.emp_no
inner join departments d on d.dept_no = de.dept_no
inner join titles t on t.emp_no = e.emp_no;
```

*Para mas información sobre vistas consulta la documentación oficial* -> [MySql Views](https://dev.mysql.com/doc/refman/8.0/en/create-view.html)



````python
# Create view 
sql_Query_view = """
CREATE  OR REPLACE VIEW v_All_Employee_Data AS
SELECT e.emp_no, birth_date,first_name, last_name, e.gender, e.hire_date, d.dept_no, d.dept_name, de.from_date as dept_from_Date, de.to_date as dept_to_date, t.title, t.from_date, t.to_date
FROM employees e
inner join dept_emp de on de.emp_no = e.emp_no
inner join departments d on d.dept_no = de.dept_no
inner join titles t on t.emp_no = e.emp_no;
"""
runSQL_Query(sql_Query_view, True)
````
